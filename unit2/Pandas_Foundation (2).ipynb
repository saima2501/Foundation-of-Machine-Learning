GroupBy in Pandas

```python
data = ({
    "Name" : ["Aman","Saurav","Karan","Shubham","Abishek",None,"Sameer"],
    "Age" : [22,23,None,22,25,24,26],
    "Marks": [85,None,88,90,56,72,None],
    "Subject": ["Maths","Physics","Maths","Chemistry","Maths","Physics","Chemistry"]
})
```

```python
df=pd.DataFrame(data)
```

```python
df.groupby('Subject')['Marks'].mean()
```

```python
df.groupby('Subject')['Name'].count()
```

```python
#Multiple Aggregation
df.groupby('Subject')['Marks'].agg(['mean','max','min'])
```

```python
df.groupby('Subject')['Marks'].agg(['mean','max','min','count'])
```

```python
df.groupby(['Subject','Age'])['Marks'].mean()
```
