```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    "Name" : ["John","Jane","Bob","Alice","Saima",None,"Saundarya","Khushi",None,"Nithya"],
    "Department": ["IT","Sales","HR","IT","Sales","Hr","IT","Sales","Hr","IT"],
    "Role": ["Analyst","Manager","Developer","engineer","Manager","Analyst","Agent","Analyst","Manager","Manager"],
    "Experience": [12,13,14,7,8,5,3,6,8,9],
    "Salary" : [50000,60000,70000,560000,650000,780000,860000,620000,450000,900000],
    "Score" :[70,85,None,78,83,None,88,47,90,None]
})
```

```python
df.groupby('Department')['Salary'].mean()
```
