```python
import numpy as np
```

```python
marks=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
```

```python
print(marks)
```

```python
print(marks.shape)
```

```python
print(marks.size)
```

```python
print(type(marks))
```

```python
second_student_marks = marks[1]
print(second_student_marks)
```

```python
third_subject_marks = marks[:, 2]
print(third_subject_marks)
```

```python
student1_marks=marks[1][1]
print(student1_marks)
```

```python
avg_per_student = np.mean(marks, axis=1)
print(avg_per_student)
```

```python
avg_per_subject=np.mean(marks,axis=0)
print(avg_per_subject)
```

```python
students_above_80 = marks[avg_per_student >= 80]
print(students_above_80)
```

```python
updated_marks=marks+5
print(updated_marks)
```

```python
status = np.where(marks > 50, 1, 0)
print(status)
```

```python
new_marks=np.where(marks < 40,0,marks)
print(new_marks)
```

```python
grade_score = np.where(avg_per_student >= 85, 10,
               np.where(avg_per_student >= 70, 8,
               np.where(avg_per_student >= 50, 6, 0)))

print(grade_score)
```
