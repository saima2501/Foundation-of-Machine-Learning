```python
avg=np.mean(arr)
print(avg)
```

```python
avg_per_hostel=np.mean(arr,axis=1)
print(avg_per_hostel)
```

```python
max_value = np.max(arr)
print(max_value)
```

```python
max_per_hostel = np.max(arr, axis=1)
print(max_per_hostel)
```

```python
total_per_hostel = np.sum(arr, axis=1)
print(total_per_hostel)

highest_hostel = np.argmax(total_per_hostel)
print( highest_hostel)
```

```python
total_per_day = np.sum(arr, axis=0)
print(total_per_day)
```

```python
highest_day_index = np.argmax(total_per_day)
print( highest_day_index)
```

```python
penalized_usage = np.where(arr > 550, arr * 1.20, arr)
print(penalized_usage)
```
